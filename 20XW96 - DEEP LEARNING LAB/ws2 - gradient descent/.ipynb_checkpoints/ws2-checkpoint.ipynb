## Gradient Descent

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(expr,xold,neta):
    flag = 1
    if not sym.is_convex(expr,x):
        flag=-1
    slope = sym.diff(expr,x)
    slope_val = slope.subs(x,xold)
    i = 10
    x_vals = [xold]
    while True:
        xnew = xold - neta*(slope_val)
        slope_val = slope.subs(x,xnew)
        x_vals.append(xnew)
        if round(xold,3) == round(xnew,3):
            break
        xold = xnew
    return x_vals
        
    
```

```python
def points(expr,x_vals):
    
    descent_points = []
    for i in x_vals:
        descent_points.append(expr.subs(x,i))
    
    return descent_points
```

```python
x = sym.symbols('x')
expression = x**2 - 4*x

x_points = np.linspace(1.4,2.6,100)
y_points = points(expression,x_points)
plt.plot(x_points,y_points)

netas = [0.25,0.5,0.75]
for neta in netas:
    x_vals = gradient_descent(expression,1.5,neta)
    gdy = points(expression,x_vals,marker = 'o')
    plt.plot(x_vals,gdy)
plt.legend(netas)
```

```python
new_expression = x*sym.log(x)
new_x_points = np.linspace(0.1,1.1,100)
new_y_points = points(new_expression,new_x_points)
plt.plot(new_x_points,new_y_points)

temp_x_vals = gradient_descent(new_expression,0.1,0.15)
new_temp_gdy = points(new_expression,temp_x_vals)
plt.plot(temp_x_vals,new_temp_gdy)
plt.show()
```
