### Problem 1

```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = sym.symbols('x,y')
eq1 = sym.Eq(x+y,6)
eq2 = sym.Eq(3*x+2*y,16)
result = sym.solve([eq1,eq2],(x,y))
print(result)
```

```python
x_vals = np.linspace(0,6,100)
y_eq1 = 6 - x_vals
y_eq2 = (16-3*x_vals)/2
plt.plot(x_vals,y_eq1,label = "x+y=6")
plt.plot(x_vals,y_eq2,label = "3x+2y=16")
plt.scatter(result[x],result[y],color="red")
plt.show()
```

### Problem 3

```python
x_val = np.linspace(-1,1,2000)
def fx2(x):
    return x**2
def fx3(x):
    return x**3
def fx(x):
    return x
def xlogx(x):
    return x*np.log(x)
def modx(x):
    return abs(x)

y1 = fx2(x_val)
y2 = fx3(x_val)
y3 = fx(x_val)
y4 = xlogx(x_val)
y5 = modx(x_val)

plt.plot(x_val,y1)
plt.plot(x_val,y2)
plt.plot(x_val,y3)
plt.plot(x_val,y4)
plt.plot(x_val,y5)
plt.show()
```
