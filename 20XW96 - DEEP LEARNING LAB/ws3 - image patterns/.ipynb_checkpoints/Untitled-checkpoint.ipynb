#### Chekerboard pattern

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros([8,8])
m,n = image.shape
for i in range(m):
    for j in range(n):
        image[i,j] = (-1)**(i+j)  # Can also use (i+j)%2 [the patter looks inverted] 
plt.imshow(image,cmap='gray')
```

This pattern shows toggling

### Bed Of Comb Function
##### 2D impulse function

```python
BoC_image = np.zeros([256,256])
BoC_image[::10,::10] = 255
plt.imshow(BoC_image,cmap='gray')
```

##### Fourier Transform
It is complex quantity <br>
Complex number => z=a+jb ==> magnitude: sqrt(a^2+b^2) and phi = inverse_tan(b/a)<br>
np.abs() -> magnitude<br>
np.angle() -> phase

```python
img = np.zeros([256,256])
img[::25,::25]= 255

#Obtain the FT of the input
IMG = np.fft.fft2(img)
IMG = np.fft.fftshift(IMG)
plt.subplot(1,2,1),plt.imshow(img,cmap='gray')
plt.subplot(1,2,2),plt.imshow(np.abs(IMG),cmap='gray')
```

#### 2D Sinusoidal grating pattern
##### f(x,y) = sin 2*pi(fx+fy) 

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
fx = 10
fy = 0
img_sin = np.sin(2*np.pi*(fx*X+fy*Y))
img_cos = np.cos(2*np.pi*(fx*X+fy*Y))
plt.imshow(img_sin,cmap='gray')
```

```python
plt.subplot(1,2,1)
plt.imshow(img_sin,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(img_cos, cmap='gray')
```

##### Frequency in an image is the change in gray level

Rapid change means high frequency <br>
Slow change means low frequency

#### Fourier transform of 2d sinusoidal wave

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
fx = 20
fy = 20
img_sin = np.sin(2*np.pi*(fx*X+fy*Y))
IMG = np.fft.fft2(img_sin)
IMG = np.fft.fftshift(IMG)
plt.subplot(1,2,1),plt.imshow(img_sin,cmap='gray')
plt.subplot(1,2,2),plt.imshow(np.abs(IMG),cmap='gray')
```
