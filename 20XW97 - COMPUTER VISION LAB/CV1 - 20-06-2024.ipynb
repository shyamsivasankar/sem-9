### Drawing a circle in a black background
###### (x-a)**2 + (y-b)**2 = r**2

```python
img2 = np.zeros([256,256])
m , n = img2.shape
r = 20
for i in range(0,m):
    for j in range(0,n):
        if(i-128)**2 + (j-128)**2 < r**2:
            img2[i,j] = 255
plt.imshow(img2,cmap='gray')
```

The center row has a pulse signal. The ends have a spike signal.

### Drawing an ellipse in a black background

##### (((x-a)**2)/(a**2)) + (((y-x)**2)/(b**2)) = r**2

```python
img3 = np.zeros([256,256])
m , n = img3.shape
r = 10
a = 10
b = 5
for i in range(m):
    for j in range(n):
        if ((i-128)**2/(a**2)) + ((j-128)**2/(b**2)) < r**2:
            img3[i,j] = 255
plt.imshow(img3,cmap='gray')
```

# Zone Plate Image

Using trignometry.

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
Z = np.cos(256*np.pi*(X**2+Y**2))
plt.imshow(Z,cmap='gray')
```

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
Z = np.cos(256*np.pi*(X**2-Y**2))
plt.imshow(Z,cmap='gray')
```

Zone plates show symmetric patterns. Cos(theta) shows even symmetry.

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
Z2 = np.sin(256*np.pi*(X**2+Y**2))
plt.imshow(Z2,cmap='gray')
```

```python
plt.plot(Z[128,0:16])
plt.plot(Z2[128,0:16])
```

```python
x = np.linspace(-128,128,256)
y = np.linspace(-128,128,256)
X,Y = np.meshgrid(x,y)
D = np.sqrt(X**2+Y**2)
D0 = 10
img = np.exp(-D**2/(2*D0**2))
plt.imshow(img,cmap='gray')
```
