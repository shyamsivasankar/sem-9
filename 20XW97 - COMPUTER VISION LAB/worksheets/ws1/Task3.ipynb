### Storing an image in different file formats

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
img = cv2.imread('D:/sem-9/20XW97 - COMPUTER VISION LAB/worksheets/ws1/cameraman.tif')
extentions = ['png','jpg','jpeg','bmp']
for i in extentions:
    cv2.imwrite('./images/cameraman.'+i,img)
```

```python
sizes = []
names = []
for filename in os.listdir('./images'):
    sizes.append(os.path.getsize('./images/'+filename))
    _, file_extension = os.path.splitext(filename)
    names.append(file_extension)
    
plt.bar(x=names,height=sizes)
plt.show()
```
