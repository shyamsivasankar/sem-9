## Quantization of Signal

Mapping larger set of value to a smaller set of value
<br>
<br>
<b>Advantage</b> : Compression
<br>
<b>Disadvantage</b> : Uniqueness is lost - loss of information

#### Terminology
<br>
1. Dynamic range - DR <br>
2. Bit Budget - b - (Amount of bits allocated to represent the information)<br>
3. Reconstruction level - L = 2^b<br>
4. Quantization step function - S = DR/L

           --------------
     --x-->|Quantization|--y-->
           --------------

##### Step 1:

````python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,1,100)
x = np.sin(2*np.pi*5*t)
plt.plot(t,x)
````

##### Step 2

````python
dr = np.max(x) - np.min(x) #Dynamic Range
b = 8 #Bit Budget
l = 2**b #Reconstruction level
q = dr/l #Quantization step size
````

##### Step 3:

````python
y = np.floor(x/q)*q
````

##### Step 4:

````python
plt.plot(t,x,'r',t,y,'k')
plt.legend(['input','output'],loc=1)
````

````python
for i in range(2,9,2):
    dr = np.max(x) - np.min(x) #Dynamic Range
    b = i #Bit Budget
    l = 2**b #Reconstruction level
    q = dr/l #Quantization step size
    y = np.floor(x/q)*q
    plt.plot(t,x,'r',t,y,'k')
    plt.legend(['input','output'],loc=1)
    plt.show()
````

## UNIFORM QUANTIZATION OF IMAGE

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
x = cv2.imread('Z:/sem 9/20XW97 - COMPUTER VISION LAB/cameraman.tif')
````

````python
dr = np.max(x) - np.min(x) #Dynamic Range
b = 4 #Bit Budget
l = 2**b #Reconstruction level
q = dr/l #Quantization step size
y = np.floor(x/q)*q
````

````python
plt.subplot(1,2,1),plt.imshow(x,cmap='gray')
plt.subplot(1,2,2),plt.imshow(y,cmap='gray')
````
