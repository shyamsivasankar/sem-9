```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3])
h = np.array([1,-1]) # High Pass Filter or change detector
y = signal.convolve(x,h)
print(y)
```

```python
t = np.linspace(0,1,100)
x = signal.square(2*np.pi*5*t)
# h = np.array([1,-1])       # High Pass Filter or change detector
h = 1/11 * np.ones(11)       # Low Pass Filter
y = signal.convolve(x,h,'same')

plt.subplot(2,1,1),plt.plot(t,x)
plt.title('Input Signal')
plt.subplot(2,1,2),plt.plot(t,y)
plt.title('Output Signal')
```

A low pass filter converts a sudden change into a slow change.

```python
# Frequency Response Of High Pass Filter
h = np.array([1,-1])
w,H = signal.freqz(h,1)
plt.plot(w,np.abs(H))
```

```python
# Frequency Response Of Low Pass Filter
h = 1/11*np.ones(11)
w,H = signal.freqz(h,1)
plt.plot(w,np.abs(H))
```

### 2-d Convolution

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

#Identity Kernal
x=cv2.imread('cameraman.tif',0)
h=np.array([[0,0,0],[0,1,0],[0,0,0]])
y=cv2.filter2D(x,-1,h)
plt.subplot(1,2,1),plt.imshow(x,cmap='gray')
plt.subplot(1,2,2),plt.imshow(y,cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
x=cv2.imread('cameraman.tif',0)
h=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
y=cv2.filter2D(x,-1,h)
plt.subplot(1,2,1),plt.imshow(x,cmap='gray')
plt.subplot(1,2,2),plt.imshow(y,cmap='gray')
```
