```python
import numpy as np
import matplotlib.pyplot as plt
```

### To Generate a sine wave

##### Step 1: Generate the independent variable

```python
t = np.linspace(0,1,100)
f = 5
```

##### Step 2: Plot the sine function sin(2*pi*f*t)

```python
x = np.sin(2*np.pi*f*t)
```

```python
plt.plot(t,x)
```

The rate of change of signal is frequency.

Therefore given that the frequency is 0 then we get a DC signal

```python
x = np.sin(2*np.pi*0*t)
plt.plot(t,x)
```

```python
x = np.cos(2*np.pi*0*t)
plt.plot(t,x)
```

### To hear a sine wave

##### sounddevice library - pip install sounddevice
##### Fs -> sampling frequency
##### f  -> signal frequency

```python
import sounddevice as sd
import numpy as np
fs = 8000 #8kHz

t = np.linspace(0,1,fs)
f = 1000 #signal frequency

x = np.sin(2*np.pi*f*t)
sd.play(x,fs)
```

#### Multiple frequency component signal

```python
f1 = 1000
f2 = 2000
f3 = 3000

def func(f,t):
    return np.sin(2*np.pi*f*t)

x1 = func(f1,t**2)
x2 = func(f2,t**2)
x3 = func(f3,t**2)
x = np.concatenate([x1,x2,x3])

sd.play(x,fs)
```

```python
plt.plot(t[0:1000],x[0:1000])
```

```python
xp3 = func(f1,t**3)
plt.plot(t[0:1000],xp3[0:1000])
```
